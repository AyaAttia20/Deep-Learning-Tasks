## `i` Implementation

```python
import numpy as np

np.random.seed = 42
```

```python
class ANN:
    def __init__(self,x,y):
      m=x.shape[1]
      self.w = np.random.rand(m, 1)
      self.b=0

    def sigmoid(self, z):
      return(1/(1 + np.exp(-z)))

    def sigmoid_derivative(self, p):
        return self.sigmoid(p) * (1 - self.sigmoid(p))


    def feedforward(self,x):
      self.u = np.dot(x,self.w) + self.b
      self.y_pred = self.sigmoid(self.u)
      return self.y_pred


    def backprop(self,x,y,lr):
      self.e=self.y_pred-y
      grad_y=self.e
      grad_u=self.sigmoid_derivative(self.u)
      grad_w=x.T

      delta_w = lr * np.dot(grad_w, grad_y * grad_u)
      delta_b = lr * np.sum(grad_y * grad_u)
      self.w -= delta_w
      self.b -= delta_b
      return self.w,self.b

    def fit(self, x, y, epochs=1000, lr=0.01):
        for epoch in range(epochs):
            self.feedforward(x)
            self.backprop(x, y, lr)

            if epoch % 10 == 0:
                loss=np.mean((y - self.y_pred)**2)


                print("Epoch: ",epoch)
                print("----------------------------------")
                print("y actual: ")
                print(y)
                print()
                print("y pred: ")
                print(self.y_pred)
                print()
                print(f"Loss: {loss:.4f}")
                print("=======*=============================*======")


```

## `i` Implementation

```python
class ANN_mult:
    def __init__(self, X, y, num_hidden_nodes):

      self.X = X
      self.y = y
      self.num_hidden_nodes = num_hidden_nodes
      self.m = X.shape[0]
      self.n = X.shape[1]
      self.W1 = np.random.rand(self.n, num_hidden_nodes)
      self.W2 = np.random.rand(num_hidden_nodes, 1)
      self.b1 = np.zeros((1, num_hidden_nodes))
      self.b2 = np.zeros((1, 1))


    def sigmoid(self, z):
      return(1/(1 + np.exp(-z)))

    def sigmoid_derivative(self, p):
        return self.sigmoid(p) * (1 - self.sigmoid(p))

    def feedforward(self,x):
        self.z1 = np.dot(x,self.W1) + self.b1
        self.p1 = self.sigmoid(self.z1)
        self.z2 = np.dot(self.p1,self.W2) + self.b2
        self.y_pred = self.sigmoid(self.z2)
        return self.y_pred


    def backprop(self, x, y, lr):

      self.e = self.y_pred - y
      grad_y = self.e

      grad_z2 = grad_y * self.sigmoid_derivative(self.z2)
      grad_W2 = np.dot(self.p1.T, grad_z2)
      grad_b2 = np.sum(grad_z2, axis=0)


      grad_p1 = np.dot(grad_z2, self.W2.T)
      grad_z1 = grad_p1 * self.sigmoid_derivative(self.z1)
      grad_W1 = np.dot(x.T, grad_z1)
      grad_b1 = np.sum(grad_z1, axis=0, keepdims=True)


      self.W1 -= lr * grad_W1
      self.b1 -= lr * grad_b1
      self.W2 -= lr * grad_W2
      self.b2 -= lr * grad_b2

      return self.W1,self.b1,self.W2,self.b2

    def fit(self, x, y, epochs=1000, lr=0.01):
        for epoch in range(epochs):
            self.feedforward(x)
            self.backprop(x, y, lr)

            if epoch % 10 == 0:
                loss=np.mean((y - self.y_pred)**2)

                print("Epoch: ",epoch)
                print("----------------------------------")
                print("y actual: ")
                print(y)
                print()
                print("y pred: ")
                print(self.y_pred)
                print()
                print(f"Loss: {loss:.4f}")
                print("=======*=============================*======")
```

## `0` Import Libraries & Load Dataset

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

np.random.seed = 42
```

```python
dataset =pd.read_csv("diabetes.csv")
dataset.head()
```

```python
X =dataset.drop("Outcome",axis=1)
y =dataset["Outcome"]
```

## `i` Shallow network


```python
model=Sequential()
model.add(Dense(5, activation="relu", input_shape=(X.shape[1],)))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer=keras.optimizers.Adam(0.01), loss="binary_crossentropy", metrics=["accuracy"])
```
